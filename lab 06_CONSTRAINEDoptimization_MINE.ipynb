# Constrained optimization

Consider the following constrained optimization problem 
\begin{equation}
\max 2x + y \quad \textrm{subject to} \quad c(x,y) = 0,
\end{equation}
where $c(x,y)=4x^2 +y^2 -8$.

1. Rewrite it as a minimization problem;
2. Verify that the point $\mathbf{x}^*=[1,2]^T$ is the solution (using the KKT necessary conditions);
3.  Write the quadratic penalty function $Q(x,y,\mu)$ and the Augmented Lagrangian function $L(x,y,\mu,\lambda)$ and plot their level curves in the square $[0,0]\times[2,4]$ and for $\mu=1$, $\mu=10$ and $Î»_1=-0.1$ and comment them;

**MINIMIZATION PROBLEM**
\begin{equation}
\min -(2x + y) \quad \textrm{subject to} \quad c(x,y) = 4x^2 +y^2 -8 = 0,
\end{equation}


```python
import numpy as np
import scipy.optimize as opt
from matplotlib import pyplot as plt
```

```python
#Quadratic penalty function
def Q(x,mu):
  return -(2*x[0]+x[1]) + mu*(4*x[0]*x[0]+x[1]*x[1]-8)
#Augmented lagrangian function, lambda = -0.1
def aL(x, mu):
  return -(2*x[0]+x[1]) + mu*(4*x[0]*x[0]+x[1]*x[1]-8) - 0.1*(4*x[0]*x[0]+x[1]*x[1]-8)
```

```python
#Plot of the level curves
x = np.linspace(0,2,1000)
y = np.linspace(0,4,1000)
X,Y = np.meshgrid(x,y)
Q1 = Q([X,Y], 1)
Q10 = Q([X,Y], 10)
L1 = aL([X,Y], 1)
L10 = aL([X,Y], 10)
C = 4*X**2 + Y**2 - 8
```

```python
fig1 = plt.figure(dpi=110)

plt.subplot(1,2,1)
plt.contourf(X,Y,Q1,20)
plt.colorbar()
plt.contour(X, Y, C, colors='r')
plt.title('Q function, mu = 1')

plt.subplot(1,2,2)
plt.contourf(X,Y,Q10,20)
plt.colorbar()
plt.contour(X, Y, C, colors='r')
plt.title('Q function, mu = 10')

fig2 = plt.figure(dpi=110)
plt.subplot(1,2,1)
plt.contourf(X,Y,L1,20)
plt.colorbar()
plt.contour(X, Y, C, colors='r')
plt.title('La function, mu = 1')

plt.subplot(1,2,2)
plt.contourf(X,Y,L10,20)
plt.colorbar()
plt.contour(X, Y, C, colors='r')
plt.title('La function, mu = 10')
```
